```python
# The continue statement skips the rest of the code
# inside a loop for the current iteration and moves the loop to the next iteration.
for i in range(5):
    if i == 2:
        continue  # Skip the rest of the loop for this iteration
    print(i)

# This loop will print 0, 1, 3, and 4.
# It skips the rest of the loop body when i is 2.
```

```python
# The break statement is used to exit a loop prematurely, breaking out of the enclosing for or while loop.
for i in range(5):
    if i == 2:
        break  # Exit the loop
    print(i)

# This loop will print 0 and 1. When i becomes 2,
# it exits the loop, so 2, 3, and 4 are not printed.
```

```python
# continue: Skip to the next iteration of the loop.
# break   : Exit the loop immediately.
```

```python
# a)
e = []
for i in x :
    if i%2 == 0 :
        e.append(i)
print(e)
```

```python
# b)
s = 0
for i in e :
    s =s+i
print(s)
```

```python
# c
s = 1
for i in e :
    s =s*i
print(s)
```
